#### Paso 1: Limpieza del dataset

```python
# 1.1 Cargar datos de entrenamiento
import pandas as pd  #  type: ignore

train_data = pd.read_csv(
    "../files/input/train_data.csv.zip",
    index_col=False,
    compression="zip",
)
train_data.head()
```

```python
# 1.2 Cargar datos de prueba
import pandas as pd  #  type: ignore

test_data = pd.read_csv(
    "../files/input/test_data.csv.zip",
    index_col=False,
    compression="zip",
)
test_data.head()
```

```python
# Tamaño de los dataset
print("datos de train:", train_data.shape)
print('datos de test:', test_data.shape)
```

```python
# - Renombre la columna "default payment next month" a "default".
train_data.rename(columns={'default payment next month': 'default'}, inplace=True)
test_data.rename(columns={'default payment next month': 'default'}, inplace=True)
```

```python
# Imprimir los nombres de las columnas de ambos DataFrames
print("Columnas de train_data:", train_data.columns.tolist())
print("Columnas de test_data:", test_data.columns.tolist())
```

```python
# 3. Eliminar la columna ID
train_data.drop(columns=['ID'], inplace=True)
test_data.drop(columns=['ID'], inplace=True)
```

```python
# Imprimir los nombres de las columnas de ambos DataFrames
print("Columnas de train_data:", train_data.columns.tolist())
print("Columnas de test_data:", test_data.columns.tolist())
```

```python
# Búsqueda de valores nulos
print("nulos en train:", train_data.isnull().sum())
print("nulos en test:", test_data.isnull().sum())
```

```python
#- Elimine los registros con informacion no disponible.
# train_data.dropna(inplace=True)
# train_data.isna().sum()
test_data.dropna(inplace=True)
test_data.isna().sum()
```

```python
# Para la columna EDUCATION, valores > 4 indican niveles superiores, 
# agrupe estos valores en la categoría "others", es decir igual a 4
train_data['EDUCATION'] = train_data['EDUCATION'].apply(lambda x: x if x <= 4 else 4)
test_data['EDUCATION'] = test_data['EDUCATION'].apply(lambda x: x if x <= 4 else 4)
```

```python
# Ver conversión de la variable EDUCATION
train_data['EDUCATION'] = train_data['EDUCATION']
test_data['EDUCATION'] = test_data['EDUCATION']

# Verificar las categorías y sus frecuencias en orden alfabético
print("Frecuencia de categorías en train_data['EDUCATION']")
print(train_data['EDUCATION'].value_counts())

print("Frecuencia de categorías en test_data['EDUCATION']")
print(test_data['EDUCATION'].value_counts())
```

#### Paso 2: División de datos

```python
# Divida los datasets en x_train, y_train, x_test, y_test
# x_train y x_test contienen  todas las columnas excepto la columna default.
# y_train y y_test contiene la variable objetivo default (pago)
x_train = train_data.drop(['default'], axis=1)
y_train = train_data['default']

x_test = test_data.drop(['default'], axis=1)
y_test = test_data['default']
```

```python
# Verificar las primeras filas de X_train, y_train, X_test, y_test
print("Primeras filas de x_train:")
print(x_train.head())

print("Primeras filas de y_train:")
print(y_train.head())

# print("Primeras filas de x_test:")
# print(x_test.head())

# print("Primeras filas de y_test:")
# print(y_test.head())
```

#### Paso 3: Crear el pipeline

```python
# Identificar variables categóricas en entrenamiento
categorical_features_train = x_train.select_dtypes(include=["object", "category"]).columns

# Identificar variables categóricas en test
categorical_features_test = x_test.select_dtypes(include=["object", "category"]).columns

# Unir ambas, test mas entrenamiento
categorical_features = list(set(categorical_features_train) & set(categorical_features_test))

# Mostrar las variables categoricas
print("Categorical features in train:", categorical_features)

```

```python
# Cree un pipeline para el modelo de clasificación. Este pipeline debe
# contener las siguientes capas:
# - Transforma las variables categoricas usando el método
#   one-hot-encoding.
# - Ajusta un modelo de bosques aleatorios (random forest).

from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Variables categóricas
categorical_features = ['SEX', 'EDUCATION', 'MARRIAGE']

# Crear el preprocesador para las variables categóricas
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_features)
    ],
    remainder='passthrough'  # Mantener las demás columnas numéricas sin cambios
)

# Crear el pipeline con RandomForestClassifier
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])

# Ajustar el pipeline con los datos de entrenamiento
pipeline.fit(x_train, y_train)

# Evaluar el modelo en el conjunto de prueba
print("Modelo entrenado. Precisión en test:", pipeline.score(x_test, y_test))
```

#### Paso 4: Optimización de hiperparámetros con validación cruzada

```python
# Optimice los hiperparametros del pipeline usando validación cruzada.
# Use 10 splits para la validación cruzada. 
# Use la función de precision balanceada para medir la precisión del modelo.

from sklearn.model_selection import GridSearchCV

# # Definición de parámetros a optimizar (ajustado para mantener pocas combinaciones y no sobrecargar)
# Definir un espacio de hiperparámetros reducido

# #Primero que utilicé y me dio 0.672  y debe ser > 0.673
# param_grid = {
#     'classifier__n_estimators': [100, 200],  # Número de árboles en el bosque
#     'classifier__max_depth': [None, 10, 20, 30],  # Profundidad máxima de los árboles
#     'classifier__min_samples_split': [2, 5, 10],  # Mínimo número de muestras para dividir un nodo
#     'classifier__min_samples_leaf': [1, 3, 5]  # Mínimo número de muestras en las hojas
# }

# min_samples_leaf=2, min_samples_split=5,
#                        n_estimators=200, random_state=42
# param_grid = {
#     'classifier__n_estimators': [10,15,20,40,60,80],  # Número de árboles en el bosque 
#     'classifier__max_depth': [None,2,5,10,15],  # Profundidad máxima de los árboles
#     'classifier__min_samples_split': [2,5,10],  # Mínimo número de muestras para dividir un nodo
#     'classifier__min_samples_leaf': [1,2,5],  # Mínimo número de muestras en las hojas
# }

param_grid = {
    'classifier__n_estimators': [80],  # Número de árboles en el bosque 
    'classifier__max_depth': [None,2,5],  # Profundidad máxima de los árboles
    'classifier__min_samples_split': [5],  # Mínimo número de muestras para dividir un nodo
    'classifier__min_samples_leaf': [2],  # Mínimo número de muestras en las hojas
}
# Usar GridSearchCV
grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=10,  # Usar validación cruzada con 10 splits
    scoring='balanced_accuracy',  # Usar precisión balanceada
    verbose=1, # como va el progreso a nivel de busqueda de hiperparametros
    n_jobs=-1  # usa todos los núcleos disponibles, mejora el rendimiento
)

grid_search.fit(x_train, y_train)
```

#### Paso 5: Guardar el modelo entrenado

```python
import pickle

# Guardar el modelo entrenado (usando el mejor estimador de RandomizedSearchCV)
with open('../files/models/model.pkl', 'wb') as file:
    pickle.dump(grid_search, file)

# import pickle
# import zipfile

# # Guardar el modelo en un archivo temporal
# with open('model.pkl', 'wb') as file:
#     pickle.dump(grid_search.best_estimator_, file)

# # Comprimir el archivo .pkl en un archivo .zip
# with zipfile.ZipFile('../files/models/model.pkl.zip', 'w', zipfile.ZIP_DEFLATED) as zipf:
#     zipf.write('model.pkl')
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, balanced_accuracy_score, confusion_matrix

# Suponiendo que el mejor modelo es el de grid_search.best_estimator_
model = grid_search.best_estimator_

# Hacer predicciones sobre el conjunto de entrenamiento
y_train_pred = model.predict(x_train)

# Hacer predicciones sobre el conjunto de prueba
y_test_pred = model.predict(x_test)

# Calcular las métricas para el conjunto de entrenamiento
train_precision = precision_score(y_train, y_train_pred)
train_balanced_accuracy = balanced_accuracy_score(y_train, y_train_pred)
train_recall = recall_score(y_train, y_train_pred)
train_f1 = f1_score(y_train, y_train_pred)

# Calcular las métricas para el conjunto de prueba
test_precision = precision_score(y_test, y_test_pred)
test_balanced_accuracy = balanced_accuracy_score(y_test, y_test_pred)
test_recall = recall_score(y_test, y_test_pred)
test_f1 = f1_score(y_test, y_test_pred)

# Mostrar las métricas de entrenamiento
print("Métricas de entrenamiento:")
print(f"Precision: {train_precision}")
print(f"Balanced Accuracy: {train_balanced_accuracy}")
print(f"Recall: {train_recall}")
print(f"F1 Score: {train_f1}")

# Mostrar las métricas  de prueba
print("\nMétricas de prueba")
print(f"Precision: {test_precision}")
print(f"Balanced Accuracy: {test_balanced_accuracy}")
print(f"Recall: {test_recall}")
print(f"F1 Score: {test_f1}")

# Calcular y mostrar la matriz de confusión para el conjunto de entrenamiento
train_cm = confusion_matrix(y_train, y_train_pred)
print("\nConfusion Matrix (Training):")
print(train_cm)

# Calcular y mostrar la matriz de confusión para el conjunto de prueba
test_cm = confusion_matrix(y_test, y_test_pred)
print("\nConfusion Matrix (Test):")
print(test_cm)

```

#### Pasos 6 y 7 : Calcular métricas, calcular matrices de confusión y guardar en JSON

```python
# Paso 6.
# Calcule las metricas de precision, precision balanceada, recall,
# y f1-score para los conjuntos de entrenamiento y prueba.
# Guardelas en el archivo files/output/metrics.json. Cada fila
# del archivo es un diccionario con las metricas de un modelo.
# Este diccionario tiene un campo para indicar si es el conjunto
# de entrenamiento o prueba. Por ejemplo:
#
# {'dataset': 'train', 'precision': 0.8, 'balanced_accuracy': 0.7, 'recall': 0.9, 'f1_score': 0.85}
# {'dataset': 'test', 'precision': 0.7, 'balanced_accuracy': 0.6, 'recall': 0.8, 'f1_score': 0.75}
#
#
# Paso 7.
# Calcule las matrices de confusion para los conjuntos de entrenamiento y
# prueba. Guardelas en el archivo files/output/metrics.json. Cada fila
# del archivo es un diccionario con las metricas de un modelo.
# de entrenamiento o prueba. Por ejemplo:
#
# {'type': 'cm_matrix', 'dataset': 'train', 'true_0': {"predicted_0": 15562, "predicte_1": 666}, 'true_1': {"predicted_0": 3333, "predicted_1": 1444}}
# {'type': 'cm_matrix', 'dataset': 'test', 'true_0': {"predicted_0": 15562, "predicte_1": 650}, 'true_1': {"predicted_0": 2490, "predicted_1": 1420}}
#


from sklearn.metrics import classification_report, balanced_accuracy_score, confusion_matrix
import json

# Función para calcular las métricas de precisión, recall, etc.
def compute_metrics(model, X, y, dataset):
    y_pred = model.predict(X)
    metrics = classification_report(y, y_pred, output_dict=True)
    
    # formato de json
    results = {
        'type': 'metrics',
        'dataset': dataset,
        'precision': metrics['1']['precision'],
        'balanced_accuracy': balanced_accuracy_score(y, y_pred),
        'recall': metrics['1']['recall'],
        'f1_score': metrics['1']['f1-score']
    }
    return results

# Función para calcular la matriz de confusión
def compute_confusion_matrix(model, X, y, dataset):
    cm = confusion_matrix(y, model.predict(X))

    # formato de json
    return {
        'type': 'cm_matrix',
        'dataset': dataset,
        'true_0': {'predicted_0': int(cm[0][0]), 'predicted_1': int(cm[0][1])},
        'true_1': {'predicted_0': int(cm[1][0]), 'predicted_1': int(cm[1][1])}
    }

# Calcula las métricas para entrenamiento y prueba
metrics_list = [
    compute_metrics(grid_search.best_estimator_, x_train, y_train, 'train'),
    compute_metrics(grid_search.best_estimator_, x_test, y_test, 'test')
]

# Calcula la matriz de confusión para entrenamiento y prueba
cm_train = compute_confusion_matrix(grid_search.best_estimator_, x_train, y_train, 'train')
cm_test = compute_confusion_matrix(grid_search.best_estimator_, x_test, y_test, 'test')

# Crear una lista con todas las métricas y matrices
all_results = metrics_list + [cm_train, cm_test]

#Guardar todas las métricas y matrices en el archivo 'metrics.json'
with open('../files/output/metrics.json', 'w') as file:
    for result in all_results:
        file.write(json.dumps(result) + '\n')
```
